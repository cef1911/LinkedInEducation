### Demo Pie Chart

```python
# https://datatofish.com/plot-dataframe-pandas/
import pandas as pd
import matplotlib.pyplot as plt

data = {'Tasks': [300,500,700]}
df = pd.DataFrame(data,columns=['Tasks'],index = ['Tasks Pending','Tasks Ongoing','Tasks Completed'])

df.plot.pie(y='Tasks',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
plt.show()
```

### Certifications Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Certifications': [9,3,5,1]}
dfcerts = pd.DataFrame(data,columns=['Certifications'],index = ['Coursera Certs','DataCamp Certs','SAS Certified Certs','Google Analytics Certs'])

dfcerts.plot.pie(y='Certifications',figsize=(5, 5),autopct='%1.1f%%', startangle=30)
plt.show()
```
